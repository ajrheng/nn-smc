```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_context("paper")

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
nn_file = np.load("./files/results_nn.npz")
nn_omegas = nn_file['true_omegas']
nn_data = nn_file['data']
```

```python
nn_data_sq = ( np.array(nn_data) - np.array(nn_omegas).reshape(-1,1)) ** 2
nn_data_mean = np.mean(nn_data_sq, axis=0)
nn_data_median = np.median(nn_data_sq, axis =0)
```

```python
nn_iters_arr = np.arange(len(nn_data_mean))

plt.plot(nn_iters_arr, nn_data_mean, label='Mean')
plt.plot(nn_iters_arr, nn_data_median, label='Median')
plt.yscale('log')
plt.legend(fontsize=13)
plt.xlabel('Iterations')
plt.ylabel('Squared Error')
plt.title("NN Performance", fontsize=14)
plt.show()
```
